```python
import numpy as np
import pandas as pd 
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['AAPL', 'AMZN','MSFT','V','ADBE', 'MA', 'NFLX','GOOG', 'FB','NVDA']

portfolio_data = pd.DataFrame()

for t in tickers:
    portfolio_data[t]= wb.DataReader(t, data_source= 'yahoo', start='2000-01-01')['Adj Close']
```

```python
portfolio_data.tail()
```

```python
portfolio_return = (portfolio_data / portfolio_data.shift(1)) -1
portfolio_return
```

```python
portfolio_return.plot(figsize=(15, 8))
plt.show()
```

```python
avg_return_a  = portfolio_return.mean()*252*100
avg_return_a
```

```python
Total_avg_return_annually =avg_return_a.mean()
Total_avg_return_annually
```

```python
weights = np.array([0.10, 0.10 , 0.10 , 0.10, 0.10, 0.10, 0.10 ,0.10, 0.10 ,0.10])
```

```python
pfolio_var = np.dot(weights.T, np.dot(portfolio_return.cov()*250 ,weights))
pfolio_var
```

```python
# portfolio Volatility 
```

```python
pfolio_vol = (np.dot(weights.T, np.dot(portfolio_return.cov()*250 ,weights))) ** 0.5
```
